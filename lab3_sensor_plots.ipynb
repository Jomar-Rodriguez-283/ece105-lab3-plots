```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = 9461

import numpy as np
import matplotlib.pyplot as plt

# Set up random number generator with seed
rng = np.random.default_rng(seed=9461)

# Generate timestamps: evenly spaced from 0 to 10 seconds
timestamps = np.linspace(0, 10, 200)

# Generate sensor readings
sensor_a = rng.normal(loc=25, scale=3, size=200)
sensor_b = rng.normal(loc=27, scale=4.5, size=200)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
# Scatter plot implementation
import matplotlib.pyplot as plt
import numpy as np
# Ensure arrays are available
assert 'sensor_a' in globals() and 'sensor_b' in globals() and 'timestamps' in globals(), 'Run the first cell to create sensor arrays'

plt.figure(figsize=(8, 4))
plt.scatter(timestamps, sensor_a, c='tab:blue', label='Sensor A', alpha=0.7)
plt.scatter(timestamps, sensor_b, c='tab:orange', label='Sensor B', alpha=0.7)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor readings vs Time')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Histogram comparison of sensor readings
import numpy as np
import matplotlib.pyplot as plt
# Ensure arrays exist
assert 'sensor_a' in globals() and 'sensor_b' in globals(), 'Run the first cell to create arrays'

# Remove NaNs for histogramming
a = sensor_a[~np.isnan(sensor_a)]
b = sensor_b[~np.isnan(sensor_b)]

# Shared bins over combined range
bins = np.linspace(min(a.min(), b.min()), max(a.max(), b.max()), 30)

plt.figure(figsize=(8,4))
plt.hist(a, bins=bins, alpha=0.6, label=f'Sensor A (n={a.size})', color='tab:blue', edgecolor='black')
plt.hist(b, bins=bins, alpha=0.6, label=f'Sensor B (n={b.size})', color='tab:orange', edgecolor='black')

# Mark means
plt.axvline(np.mean(a), color='navy', linestyle='--', linewidth=1, label='A mean')
plt.axvline(np.mean(b), color='darkorange', linestyle='--', linewidth=1, label='B mean')

plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Histogram of sensor temperatures')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
import matplotlib.pyplot as plt
import numpy as np

assert 'sensor_a' in globals() and 'sensor_b' in globals(), 'Run the first cell to create sensor arrays'

# Remove NaNs if present
sensor_a_clean = sensor_a[~np.isnan(sensor_a)]
sensor_b_clean = sensor_b[~np.isnan(sensor_b)]

overall_mean = np.mean(np.concatenate([sensor_a_clean, sensor_b_clean]))

plt.figure(figsize=(8, 5))
plt.boxplot([sensor_a_clean, sensor_b_clean], labels=['Sensor A', 'Sensor B'], patch_artist=True,
            boxprops=dict(facecolor='tab:blue', alpha=0.4), medianprops=dict(color='black'))
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=1.5, label=f'Overall mean ({overall_mean:.2f} °C)')
plt.ylabel('Temperature (deg C)')
plt.title('Box plot comparison of Sensor A and Sensor B')
plt.legend()
plt.grid(alpha=0.3, axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
